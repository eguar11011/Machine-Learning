```python
import pandas as pd
import os
import json
```

```python
# ruta del dataset
base_dir = 'mmimdb'
dataset_dir = os.path.join(base_dir, 'dataset')


# Abre el archivo JSON
with open(os.path.join(base_dir,'split.json') ) as archivo:
    # Carga el archivo JSON en una estructura de datos de Python
    ids = json.load(archivo)
```

```python
ids["test"]
```

```python
drama = list()
horror = list()
i = 0

for file in ids:
    for i in ids[file]:
        # Abrir el archivo json
        with open( os.path.join(dataset_dir, i+".json" )) as archivo:
            sample = json.load(archivo)
        # Guardar el id para Drama
        if 'Drama' in sample["genres"] and  "Horror" not in sample["genres"] and len(drama)< 2000:
            drama.append( i+".json")
        # Guardar el id para Horror
        if 'Horror' in sample["genres"] and  "Drama" not in sample["genres"] and len(horror)< 2000:
            horror.append( i+".json")

        if  len(drama) == 2000 and  len(horror) == 2000:
            break
```

```python
print("Tamaño de Drama y Horror:" ,len(drama), len(horror))
```
